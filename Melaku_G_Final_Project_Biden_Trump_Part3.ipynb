__Import__ the tweets

```python
#  Import Biden"s tweet data  as a dataframe

import pandas as pd

biden = pd.read_csv("JoeBiden.csv")
biden.head(3)
```

```python
# see how many rows and column the dataframe has

biden.shape
```

```python
#  Import Trumps tweet data  as a dataframe

trump = pd.read_csv("DonaldTrump.csv")
trump.head(3) 

```

```python
# see how many rows and column Trump dataframe has

trump.shape
```

```python
# First let's change the column heading from 'text' to tweet

biden.rename(columns={'text':'tweet'}, inplace=True)
biden.head(2)
```

```python
# First let's change the column heading from 'text' to tweet (trump)

trump.rename(columns={'text':'tweet'}, inplace=True)
trump.head(2)
```

```python
# change tweets to LOWER Case
biden['tweet'] = biden['tweet'].str.lower()
trump['tweet'] = trump['tweet'].str.lower()
```

```python
#Biden
# Create a column for charcounts
biden['charcounts']=biden['tweet'].str.len()

#  create a column "wordcount"
biden['wordcount'] = biden['tweet'].str.split().str.len()

# create a column "avgwordlen"
biden['avgwordlen']=biden['charcounts'] / biden['wordcount']

#creat a column "hash "
biden['hash'] = biden.tweet.str.findall("(?<=#)\w+")

#create a column "mentions"
biden['mentions']=biden.tweet.str.findall("(?<=@)\w+")

#average number of hashtags lengt per tweet
biden['AvgHashtag']=biden['hash'].str.len()

#Trump
# Create a column for charcounts
trump['charcounts']=trump['tweet'].str.len()

#  create a column "wordcount"
trump['wordcount'] = trump['tweet'].str.split().str.len()

# create a column "avgwordlen"
trump['avgwordlen']=trump['charcounts'] / biden['wordcount']

#creat a column "hash "
trump['hash'] = trump.tweet.str.findall("(?<=#)\w+")

#create a column "mentions"
trump['mentions']=trump.tweet.str.findall("(?<=@)\w+")

#average number of hashtags lengt per tweet
trump['AvgHashtag']=trump['hash'].str.len()

```

### Word Frequency

```python
# using collections python library to count
from collections import Counter

#Biden
# Get counts by word in list
list(Counter(" ".join(biden['tweet']).split(" ")).items())

# Store list 
fdist = list(Counter(" ".join(biden['tweet']).split(" ")).items())

# get the first 15
fdist[:15]
```

```python
# Top 10 words 

# Biden
freq_wordcounts= Counter(fdist)
freq_wordcounts.most_common(10)
```

```python
#Trump
# Get counts by word in list
list(Counter(" ".join(trump['tweet']).split(" ")).items())

# Store list 
fdist = list(Counter(" ".join(trump['tweet']).split(" ")).items())

# get the first 15
fdist[:15]
```

```python
# Top 10 words 

# Trump
freq_wordcounts= Counter(fdist)
freq_wordcounts.most_common(10)
```

### Vader Sentiment Analysis

```python
#  Get Vader Lexicon if needed.
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Run Sentiment Analysis

# Biden
vs = SentimentIntensityAnalyzer()

#Get Scores and Put into Dataframe
biden['score'] = biden['tweet'].apply(lambda x: vs.polarity_scores(x)['compound'])
biden.head()
```

```python
# Run Sentiment Analysis

# Trump
vs = SentimentIntensityAnalyzer()

#Get Scores and Put into Dataframe
trump['score'] = trump['tweet'].apply(lambda x: vs.polarity_scores(x)['compound'])
trump.head()
```

```python
# Mean sentiment score Biden
biden['score'].mean()
```

```python
# Mean sentiment score trump
trump['score'].mean()
```
